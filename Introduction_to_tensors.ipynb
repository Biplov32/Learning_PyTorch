## Pytorch reproducblity (trying to take random out of random)
* To reduce the randomness in neural networks in pytorch comes concept of  **random seed**
Essentially what random seed does is "flavour" the randomness

```python
#create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Lets make some random but reproducable tensors
# set the random seed
random_seed = 32
torch.manual_seed(random_seed)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(random_seed)
random_tensor_D = torch.rand(3, 4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```

## Putting tensors (and models) on the GPU


```python
#Create a tensor(default on the cpu)
tensor = torch.tensor([1,2,3], device = "cpu")
print(tensor, tensor.device)
```

```python
#move tensor to the gpu if available
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

## Moving tensors back to the cpu

```python
# if tensor is on GPU , can't transfrom it to Numpy
tensor_on_gpu.numpy()
```

```python
#TO fix the GPU tensor with Numpy issue , we can first set it to the CPU
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu
```
